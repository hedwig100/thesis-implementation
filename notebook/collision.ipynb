```python
def compute_isogeny(E, R, e):
    """
    Computes E/<R> where the order of R is 2^e.
    """
    for i in range(e - 1, -1, -1):
        T = (1 << i)*R
        phi = E.isogeny(T)
        R = phi(R)
        E = phi.codomain()
    return E
```
